```python
#Ramen-Ratings
#An attempt to familiarize myself with Pandas, and possibly learn something about ramen noodles

#Initialize modules
import pandas
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (25, 10)

#Import data
ramen_ratings = pandas.read_csv("ramen-ratings.csv", index_col="review")
ramen_ratings = ramen_ratings.sort_values(by=['review']) #Because the list was unsorted
ramen_ratings['stars'] = pandas.to_numeric(ramen_ratings['stars'], errors='coerce') #Convert strings to floats for the stars rating

```

```python
#Lets start to analyze the data.
#Lets look at it first.
ramen_ratings
```

```python
#Next, let's look at the number of ramen reviews per country
countries = ramen_ratings["country"].value_counts()
```

```python
countries
```

```python
#There are a lot of places to go from here
#To start, we'll make a list of just 5 star ramens.
is_5_star = ramen_ratings['stars'] == 5
five_stars = ramen_ratings[is_5_star]
```

```python
five_stars
```

```python
#Now lets count those up by country.
countries_5_stars = five_stars['country'].value_counts()
```

```python
#And look at them
countries_5_stars
```

```python
#Let's turn that into a list we can use and do some ratio comparisons.
country_list = five_stars['country'].unique()
```

```python
#Let's peek
country_list
```

```python
#Here we declare some dictionaries to keep track of our ratios
five_star_ratio_country = {} #Ratio of 5-star ramen in that country vs that country's number of submitted ramen
five_star_ratio_overall = {} #Ratio of 5-star ramen in that country vs The total number of 5 star ramen
overall_ratio = {}           #Ratio of Country submitted ramen vs overall total reviewed
overall_points = {}          #If all other ratios are treated equally as worth 1 point, the total number of points in essence to try a ramen from that country.

#Totals 
total_reviews = countries.sum()
total_5_reviews = countries_5_stars.sum()

#Done as a for loop to go through each country in the list above, updating each dictionary with the name of the country as the key
for each in country_list:
    total_count = countries[each]                       #Define a word variable to each countries total count of ramen reviews
    five_count = countries_5_stars[each]                #word var for each countries 5 star reviews
    ratio_country = five_count / total_count            #ratio_country is the ratio of 5-star reviews against that countries total reviews
    ratio_overall = total_count / total_reviews         #ratio overall is the ratio of that country vs the overall total
    ratio_five_overall = five_count / total_reviews     #ratio five overall is the ratio of 5 star reviews to overall total
    
    #Assign the dictionary keys
    five_star_ratio_country[each] = ratio_country      
    five_star_ratio_overall[each] = ratio_five_overall
    overall_ratio[each] = ratio_overall
    overall_points[each] = ratio_country + ratio_five_overall + ratio_overall 
```

```python
#define a function to display our dictionaries as barplots
def barplot(input_dictionary):
    plt.bar(range(len(input_dictionary)),list(input_dictionary.values()), align='center')
    plt.xticks(range(len(input_dictionary)),list(input_dictionary.keys()))
    plt.show() 
    #credit: stack exchange
```

```python
#Comparison 1
barplot(five_star_ratio_country)
#This is the ratio of 5 star reviews out of the total number of ramen reviews in that country

#It looks like Cambodia, Malaysia, Sarawak, and Singapore, and Myanmar are the top 6.
```

```python
#But that's not the whole picture.
barplot(five_star_ratio_overall)
#This is the ratio of five star ramens from that country to the total number of submitted ramen in total
#Japan shows a clear lead here.
```

```python
barplot(overall_ratio)
#Here's the overall numer of ramen submitted per country vs the total number of ramen reviews available
#Again, Japan with a good lead.
```

```python
#Now if we treat each of these ratios as worth one point , who comes out ahead?
overall_points
```

```python
barplot(overall_points)
#Now the Highest is Malaysia!
```

```python
highest_country = max(overall_points, key=overall_points.get)
print(highest_country,overall_points[highest_country])
```

```python
#However, do these points mean anything?

#Despite having the most ramen reviews by country, Japan comes out 4th in terms of points, behind Malaysia, Cambodia, and Singapore. 
#The "Points" Here are really just the ratios added together.
#Now its fun to go back and take another look at the data with this lens in mind. 

#Looking at our first graph, the ratio of 5 star reviews vs reviewed ramen in that country, puts Malaysia second overall
#But our next graph hints us to refer back to the actual data, where we see we have a few possibly artifically inflated countries like Cambodia, Sarawak and Myanmar.

#No data source is perfect, and this one only has a few reviews from those countries. 
#This data set also has a few mismatches that I should go back and clean up, like 'United States' and 'US', and arguably 'China', and 'Hong Kong'

#Myanmar leads the pack with its 4 5-star reviewed ramens out of 14 total offered. That casts a pretty good light on the whole country
#Cambodia is next on the first chart with a 2 5-star out of 5 offered ratio. That's starting to look like too small of a sample size to really count.
#Sarawak is an extreme at 1:3, and to be honest, I've never heard of it. 

#So should we trust the point system? 
# For my use, this analysis might influence my next ramen purchase at the store, based on country alone, then my preference of flavor.
# - Getting a ramen from Malaysia doesn't seem like a bad item. Going down the list to Cambodia could be risky, 
# - because I'm willing to bet there are more than 5 major ramen brands in Cambodia.
# - Singapore and Japan seem like easily safe bets, having both a good ratio and a large number of submitted reviews. 

#We can really look at the data in a few more ways, both by brand, and considering the other reivew numbers besides just 5.
#More to come. 
```

```python
def brand_and_country(brand,country,rating):
    brand_list = ramen_ratings['brand'] == brand
    country_list = ramen_ratings['country'] == country
    rating_list = ramen_ratings['stars'] == rating
    return ramen_ratings[brand_list & country_list & rating_list]
```

```python
Nissin_Malaysia = brand_and_country('Nissin','Japan',5)
Nissin_Malaysia
```

```python
ratio_brand_overall = {}
ratio_five_star_brand_c = {}

for brand in top_five:
    for country in top_5_points:
        df = brand_and_country(brand,country,5)
        
        total_count = countries[country]                       #Define a word variable to each countries total count of ramen reviews
        five_count = countries_5_stars[country]                #word var for each countries 5 star reviews
        total_brand = brand_count[brand]
        five_brand = five_star_brand_count[brand]
        five_star_brand_country = df['brand'].count()          #Just to get 1 number representing the list
        
        ratio_five_star_brand_c[brand,country] = five_star_brand_country / five_brand
    ratio_brand_overall[brand] = total_brand / total_reviews
```
